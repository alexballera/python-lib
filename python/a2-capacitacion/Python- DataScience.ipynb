Define una función que calcule y devuelva el área de un círculo dado su radio

```python
import math
def area_circulo(r):
    return round(math.pi*r**2, 2)

area_circulo(15)
```

Define una función que calcule y devuelva el precio total de un producto dado su precio base y un impuesto (por defecto del 16%)

```python
def precio_total(precio, t=16):
    return round(precio*(1 + t/100), 2)

precio_total(100)
```

Define una función que reciba un número variable de argumentos y devuelva el promedio de dichos números.

```python
def promedio(*numeros):
    if (len(numeros) == 0):
        return 0
    
    return sum(numeros) / len(numeros)

print(promedio(2,3,4,5,6))
print(promedio(2,3,4,9,12,6))
print(promedio(22,35,4,5,6))
```

Crear un gráfico de dispersión con las "horas de entrenamiento" en el eje x y el "precio" en el eje y

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x=futbolistas['Horas_Entrenando'], y=futbolistas['Precio'], alpha=0.6)
ax.set_xlabel('Horas entrenando')
ax.set_ylabel('Precio')
ax.grid(True)
ax.set_title('Precio vs Horas Entrenando')
fig.tight_layout()
plt.show()
```

Crear un gráfico de barras con las horas de entrenamiento según las prácticas y si fuman o no

```python
entrenamiento.plot.bar()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Obtener los índices y preparar los datos
categorias = entrenamiento.index
x_pos = np.arange(len(categorias))
width = 0.35  # Ancho de las barras

# Crear las barras manualmente con ax.bar()
bars1 = ax.bar(x_pos - width/2, entrenamiento[False], width, 
               label='No Fuma', color='#2ecc71', alpha=0.8, edgecolor='black')
bars2 = ax.bar(x_pos + width/2, entrenamiento[True], width,
               label='Fuma', color='#e74c3c', alpha=0.8, edgecolor='black')

# Personalizar el gráfico
ax.set_title('Horas de Entrenamiento según Prácticas y si Fuman', fontsize=14, fontweight='bold')
ax.set_xlabel('Prácticas', fontsize=12)
ax.set_ylabel('Horas de Entrenamiento', fontsize=12)
ax.set_xticks(x_pos)
ax.set_xticklabels(categorias, rotation=0)
ax.legend(title='Fuma', loc='upper right')
ax.grid(axis='y', alpha=0.3, linestyle='--')

fig.tight_layout()
plt.show()
```
