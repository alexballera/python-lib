# Broadcasting (operaciones)

```python
import numpy as np

# Crear arrays para demostrar broadcasting
a = np.array([1, 2, 3, 4])
b = np.array([[1], [2], [3]])

print("Forma del array a:", a.shape)
print("Array a:")
print(a)
print("\nForma del array b:", b.shape) 
print("Array b:")
print(b)

# Ejemplo de broadcasting - suma de arrays con diferentes formas
print("\n=== SUMA CON BROADCASTING ===")
resultado_suma = a + b
print("Resultado de a + b (broadcasting):")
print("Forma del resultado:", resultado_suma.shape)
print(resultado_suma)

# Ejemplo de broadcasting - multiplicación de arrays con diferentes formas
print("\n=== MULTIPLICACIÓN CON BROADCASTING ===")
resultado_mult = a * b
print("Resultado de a * b (broadcasting):")
print("Forma del resultado:", resultado_mult.shape)
print(resultado_mult)

# Ejemplo adicional con escalares
print("\n=== BROADCASTING CON ESCALARES ===")
escalar = 5
suma_escalar = a + escalar
mult_escalar = a * escalar
print(f"a + {escalar} =", suma_escalar)
print(f"a * {escalar} =", mult_escalar)
```
